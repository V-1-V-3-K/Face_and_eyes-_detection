# Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Import Image

```python
image = cv2.imread("Images/face_eyes.jpg")
plt.imshow(image)
```

```python
fix_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

# Create Classifier

```python
eye_classifier = cv2.CascadeClassifier("haar-cascade-files-master/haarcascade_eye.xml")
```

# Finding and Making Rectangles over eyes

```python
def eye_detect(fix_img):
    eye_rect = eye_classifier.detectMultiScale(fix_img)
    for (x,y,w,h) in eye_rect:
        cv2.rectangle(fix_img,
                     (x,y),
                     (x+w,y+h),
                     (255,0,0),
                     10)
    return fix_img
```

```python
result = eye_detect(fix_img)
plt.imshow(result)
```
