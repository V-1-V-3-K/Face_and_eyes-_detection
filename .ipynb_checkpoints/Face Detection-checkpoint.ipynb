# Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Import image

```python
img = cv2.imread("Images/face_eyes.jpg")
plt.imshow(img)
```

```python
img_fix = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_fix)
```

# Classifier

```python
face_classifier = cv2.CascadeClassifier("haar-cascade-files-master/haarcascade_frontalface_default.xml")
```

# Detecting Faces

```python
faces = face_classifier.detectMultiScale(img, 1.3, 5)
```

```python
if faces is ():
    print("no faces detected")
```

```python
def detect_face(img_fix):
    face_rect = face_classifier.detectMultiScale(img_fix)
    for (x,y,w,h) in face_rect:
        cv2.rectangle(img_fix,
                     (x,y),
                     (x+w,y+h),
                     (0,255,0),
                     10)
    return img_fix
```

```python
result = detect_face(img_fix)
plt.imshow(result)
```
